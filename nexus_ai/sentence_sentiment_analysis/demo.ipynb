```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
aspects = []
for sentence in reviews:
    doc = nlp(sentence)
    descriptive_term = ''
    target = ''
    for token in doc:
        if token.dep_ == 'nsubj' and token.pos_ == 'NOUN':
            target = token.text
        if token.pos_ == 'ADJ':
            prepend = ''
            for child in token.children:
                if child.pos_ != 'ADV':
                    continue
                prepend += child.text + ' '
            descriptive_term = prepend + token.text
    aspects.append({'aspect': target,
        'description': descriptive_term})
print(aspects)
```

```python
for idx,aspect in enumerate(aspects):
    if aspect['description'] == '' or aspect['aspect'] == '':
        del aspects[idx]
```

## using string one at a time:

```python
%%time
for idx,aspect in enumerate(aspects):
    if aspect['description'] == '' or aspect['aspect'] == '':
        del aspects[idx]
for aspect in aspects:
    aspect['sentiment'] = predict(reviews=aspect['description'], gpu=True)
```

```python
ascpets_pd = pd.DataFrame(aspects)
ascpets_pd
```
